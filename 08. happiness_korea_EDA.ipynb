```python
# 필요한 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 시드 고정
np.random.seed(42)
```

```python
# 엑셀 파일 읽기
df = pd.read_excel('happiness_korea.xlsx', engine='openpyxl')
df.head()
```

```python
# 데이터 형상 파악하기
df.info()
```

```python
df.describe()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
# Correlation Analysis

corr = df.corr()
sns.set(font_scale=3)
plt.figure(figsize=(25, 25));
ax = sns.heatmap(corr,
            vmax=0.8,
            linewidths=0.1,
            square=True,
            annot=True,
            cmap='Blues',
            annot_kws={"size": 30})
ax.tick_params(labelsize=30)

plt.title('Feature Correlation of happiness_korea');
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
```

```python
# x축과 y축을 변환하여 데이터프레임 정렬
sorted_df = df.nlargest(10, 'Happiness').sort_values(by='Happiness', ascending=False)

# 그래프 그리기 (x축과 y축을 변환하여 사용)
sns.barplot(x='City/District', y='Happiness', data=sorted_df, palette='crest', ci=None, dodge=False)

# 변수 이름 조정
plt.xlabel('City/District', fontsize=10)  # x축 변수 이름 조정
plt.ylabel('Happiness Score', fontsize=10)  # y축 변수 이름 조정

# 그래프에 대한 글꼴 크기 조절
plt.xticks(fontsize=10)  # x축 눈금의 글꼴 크기 조절
plt.yticks(fontsize=10)  # y축 눈금의 글꼴 크기 조절

# 그래프 출력
plt.show()
```
